```python
# merge attendance with members
df1 = pd.merge(df_attendance, df_members, how='left', left_on='link_to_member', right_on='member_id')

# merge df1 with major
df2 = pd.merge(df1, df_major, how='left', left_on='link_to_major', right_on='major_id')

# # merge df2 with zip_code
df3 = pd.merge(df2, df_zip, how='left', left_on='zip', right_on='zip_code')

# # Merge df3 with income
df4 = pd.merge(df3, df_income, how='left', on='link_to_member')

# # merge df4 with expense
df5 = pd.merge(df4, df_expense, how='left', on='link_to_member')

# # merge df5 with event
df6 = pd.merge(df5, df_event, how='left', left_on='link_to_event', right_on='event_id')

# # Merge df6 with budget
df7 = pd.merge(df6, df_budget, how='left', left_on='link_to_event', right_on='link_to_event')

# #  Drop redundant columns
# df7.drop(columns = ['link_to_member', 'link_to_major', 'zip', 'link_to_event'], inplace=True)

# # # # Save the  dataset
df7.to_csv('consolidated_dataset.csv', index=False)
```
